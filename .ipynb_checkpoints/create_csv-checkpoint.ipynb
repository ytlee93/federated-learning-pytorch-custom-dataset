```python
import os 
import numpy as np
import csv
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
# get cwd parent directory
path = os.path.dirname(cwd)
print(path)
arr = os.listdir(path+"\FL_DP_covid\FinalCovid19Dataset")
print(arr)
```

```python
train_dir = path+"\FL_DP_covid\FinalCovid19Dataset\\train"
```

```python
dirs = os.listdir(train_dir)
```

```python
train_arr = []
```

```python
for i in dirs:
    dir_list = []
    print(train_dir+"\\"+i)
    dir_list = os.listdir(train_dir+"\\"+i)
    print(len(dir_list))
    for j in dir_list:
        img_path = train_dir+"\\"+i+"\\"+j
        train_arr.append([img_path, int(i)])
        
print(train_arr)
```

```python
os.getcwd()
```

```python
with open('train.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    for i in train_arr:
        writer.writerow(i)
```

```python
test_dir = path+"\FL_DP_covid\FinalCovid19Dataset\\test"
```

```python
test_arr = []
```

```python
for i in dirs:
    dir_list = []
    print(test_dir+"\\"+i)
    dir_list = os.listdir(test_dir+"\\"+i)
    print(len(dir_list))
    for j in dir_list:
        img_path = test_dir+"\\"+i+"\\"+j
        test_arr.append([img_path, int(i)])
        
print(test_arr)
```
